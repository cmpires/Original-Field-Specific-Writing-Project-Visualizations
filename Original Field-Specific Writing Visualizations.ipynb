```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['flower'] = iris.target
df['flower'] = df['flower'].apply(lambda x: iris.target_names[x])
df.head(5)
```

```python
sns.lineplot(x=df['sepal length (cm)'], y=df['sepal width (cm)'])
```

```python
sns.lineplot(x=df['sepal length (cm)'], y=df['sepal width (cm)'], hue=df['flower'])
```

```python
sns.scatterplot(x=df['petal length (cm)'], y=df['petal width (cm)'])
```

```python
sns.scatterplot(x=df['petal length (cm)'], y=df['petal width (cm)'], hue=df['flower'])
```

```python
sns.boxplot(x=df['sepal length (cm)'])
```

```python
sns.boxplot(x=df['sepal length (cm)'], y=df['flower'])
```

```python
sns.violinplot(x=df['flower'], y=df['petal width (cm)'])
```

```python
sns.stripplot(x=df['flower'], y=df['petal width (cm)'])
```

```python
sns.countplot(x=df['flower'])
```

```python
sns.barplot(x=df['petal width (cm)'], y=df['flower'])
```
